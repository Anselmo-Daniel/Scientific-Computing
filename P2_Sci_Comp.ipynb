# Questão 2

Temos a equação: $$ \epsilon \frac{\mathrm{d}^{2}u}{\mathrm{d}x^2} + 2(1-x^2)u + u^2 = 1 $$

<br>

$$
\frac{\mathrm{d}^{2}u}{\mathrm{d}x^2} = \frac{1 - 2(1-x^2)u - u^2}{\epsilon}
$$

<br>

Tomando $ - \frac{\mathrm{d}^2u}{\mathrm{d}x^2} = f(x)$, e uma solução como $u(x) \approx u_h(x) = \sum^N_{j = 1} c_j \phi_j(x)$, como temos $ u_h(x)$ uma aproximação para $ u(x) $, se inserirmos no problema de BVP, iremos obter uma função resíduo como:

$$ r(x) = \frac{\mathrm{d}^{2}u_h}{\mathrm{d}x^2} + f(x)  $$

<br>

Não conseguimos obter a função $r(x)$ diretamente, pois a função : $$ f(x) = \frac{ 2(1-x^2)u + u^2 - 1}{\epsilon} $$

Depende de $u$ e $u^{2}$, que são termos desconhecidos no nosso problema.

<br>

Então iremos usar uma aproximação para a função $f(x)$ usando as aproximações $u_h(x)$, que iremos chamar de $$ f_h(x) = \frac{ 2(1-x^2)u_h + u_h^2 - 1}{\epsilon} $$

<br>

Se tivessemos a função $f$ original, teriamos que:

$$ \int^1_{-1}r(x)\phi_i(x) \mathrm{d}x = 0 $$

Pois temos que a função $r(x)$ é ortogonal a $ \phi_i(x)$.

<br>

Como temos a função $f_h(x)$ ao invés da $f(x)$, temos que :

$$ \int^1_{-1}r_h(x)\phi_i(x) \mathrm{d}x \approx 0 $$

pois a função $r_h(x)$ é uma aproximação de $r(x)$.

$$ = - c_{j} \cdot (j\pi)^{2} + \frac{1}{\epsilon} \left[ 2 \cdot c_{j} - \frac{2 c_{j}}{3} + \frac{ c_{j}}{ (\pi j)^{2}} - \sum_{j \neq i} c_{j} \cdot \int_{-1}^{1} 2 x^{2} sin(j\pi x) \cdot sin(i\pi x)\mathrm{d}x \right] \approx 0 $$

Temos que dividir a equação acima em 2 casos para montar nossa matriz $ A$. O primeiro caso é para $ i = j$ e temos:

$$ A_{ij} = \frac{-2}{\epsilon}\int_{-1}^{1} x^{2} sin(j\pi x) \cdot sin(i\pi x)\mathrm{d}x  $$

<br>

Dessa forma, para obtermos a solução via F.E.M, precisamos somente resolver o sistema linear : $$ A\cdot \vec{c} \approx 0 $$

Como iremos fazer o cálculo númerico de $ \int_{-1}^{1} x^{2} sin(j\pi x) \cdot sin(i\pi x)\mathrm{d}x $, temos todas as variáveis para determinar os coeficientes $c_{j}$ que fazem o sistema linear acima ficar próximo de zero.





## Para $\epsilon = 1$

```python
n = [i for i in range(1, 100)] # numero de bases phi_j´s iremos usar
eps = 1
x_s = np.linspace(-1, 1, num = 100) # pontos do grid

def base_phi(j): # função phi_j
  def phi_j(x, j = j):
    return np.sin(j*np.pi*x)
  return phi_j

vec_base = [base_phi(j) for j in n] # todas as bases para um determinado tamanho len(n)

A = np.zeros((len(n), len(n))) # construção da matriz A

for i in range(len(n)):
    for j in range(len(n)):
        if j == i:
            A[i][j] = (-1)*((np.pi*(i+1))**2) + (2 - 2/3 + 1/((np.pi*(i+1))**2))/eps
        else:
            def integ(phi1,phi2):
                def integ_x(x):
                    return phi1(x)*phi2(x)*(x**2)
                return integ_x

            f = integ(base_phi(i),base_phi(j))
            y_i = f(x_s)

            A[i][j] = simpson(y_i,x_s) * (-2/eps)
```

```python
aprox = [1e-9 for i in range(len(n))] # matriz aproximada e igual a 0

c_j1 = np.linalg.solve(A,aprox) # Calculando o vetor c_j
c_j1
```

```python
def u(c_j, n=n ): # Construção da função aproximada u_h
    def u_h(x, n = n):
        soma = 0
        for i,j in enumerate(n):
            soma += base_phi(j)(x)*c_j[i]
        return soma
    return u_h
```

## Montando o Gráfico de $ u_{h}(x)$

```python
eixo_y_u = []

for i in x_s:
  eixo_y_u.append(u(c_j1, n)(i))

eixo_y_u = np.array(eixo_y_u)

plt.figure(figsize=(16,7))
plt.plot(x_s, eixo_y_u)
plt.title('Sol da Eq diferencial para  $\epsilon = 1$')
plt.show()
```

## Montando o Gráfico de $ f_{h}(x)$

```python
def f_h1(u_h, epsil = eps):
    def f(x):
        return (2*(1-x**2)*u_h(x) + u_h(x)**2 -1)/epsil
    return f
```

```python
eixo_y_f1 = []

for i in x_s:
  eixo_y_f1.append( f_h1(u(c_j1))(i) )

eixo_y_f1 = np.array(eixo_y_f1)
```

```python
plt.figure(figsize=(16,7))
plt.plot(x_s, eixo_y_f1)
plt.title(' $f_h(x)$ para  $\epsilon = 1$')
plt.show()
```
